```python
from transformers import pipeline

pipe = pipeline("image-classification", model="Tater86/room-classifier-v1")
```

```python
import os
from collections import defaultdict
from PIL import Image

def rename_images_with_predictions(directory, model):
    room_counts = defaultdict(int)

    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')) and filename.lower().startswith('image'):

            image_path = os.path.join(directory, filename)

            image = load_image(image_path)

            # Get image classification predictions
            predictions = model(image)

            # Get top prediction label
            top_prediction = predictions[0]['label']

            # Increment the room type count
            room_counts[top_prediction] += 1

            # Generate the new file name with an index
            new_filename = f"{top_prediction}_{room_counts[top_prediction]}.jpg"
            new_image_path = os.path.join(directory, new_filename)

            # Rename the image
            os.rename(image_path, new_image_path)
            print(f"Renamed {filename} to {new_filename}")

def load_image(image_path):
    from PIL import Image
    return Image.open(image_path)

def process_image_folders(base_directory, model):
    for root, dirs, files in os.walk(base_directory):
        for dir_name in dirs:
            directory = os.path.join(root, dir_name)
            rename_images_with_predictions(directory, model)

directory = content_url + '/images'
process_image_folders(directory, pipe)
```

```python
from collections import defaultdict

def count_room_types(base_directory):
    room_counts = defaultdict(int)

    for root, dirs, files in os.walk(base_directory):
        for filename in files:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')) and not filename.lower().startswith('composite'):
                room_type = filename.split('_')[0]
                room_counts[room_type] += 1

    return room_counts


room_counts = count_room_types(directory)


for room_type, count in room_counts.items():
    print(f"{room_type}: {count}")
```
